# Lab 4 — Linear Regression
Fitting and diagnosing a simple linear regression model with `engstats`.

```python
import engstats as es
import matplotlib.pyplot as plt
```

```python
df = es.load_dataset('concrete')
df.head()
```

```python
model = es.simple_linear_regression(df, x='water_cement', y='strength_mpa')
model.summary()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
es.plot_scatter_regression(df, x='water_cement', y='strength_mpa', ax=axes[0])
es.plot_residuals(model, ax=axes[1])
es.plot_qq(model, ax=axes[2])
plt.tight_layout()
```
